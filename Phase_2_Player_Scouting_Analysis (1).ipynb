### Generate the Scouting Radar

```python
# Define scouting parameters
params = ["Goals", "Shot Accuracy", "Dribbling", "Passing", "Physicality"]
low = [0, 0, 0, 0, 0]
high = [30, 100, 100, 100, 100]

# Stats for Salah (Red) and Isak (Black)
# Using your scraped goals and estimated scouting metrics
salah_stats = [29, 85, 90, 82, 75]
isak_stats  = [23, 78, 85, 70, 88]

# Draw the Radar
radar = Radar(params, low, high, num_rings=5, center_circle_radius=1.0)

# Create figure and axes for the radar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Draw Salah's radar
radar.draw_radar(salah_stats, ax=ax, kwargs_radar={'facecolor': '#d00027', 'alpha': 0.5})

# Draw Isak's radar on the same axes for comparison
radar.draw_radar(isak_stats, ax=ax, kwargs_radar={'facecolor': '#000000', 'alpha': 0.5})

plt.title('Scouting Comparison: Salah vs Isak', fontsize=16)
plt.show()
```

### Financial Scouting Analysis

```python
data = {
    'Player': ['Mohamed Salah', 'Alexander Isak'],
    'Goals': [29, 23],
    'Value_M': [65, 75],      # Estimated Market Value in Millions
    'Weekly_Wage_K': [350, 120] # Weekly Wage in Thousands
}

df_scout_finance = pd.DataFrame(data)

# Calculate Efficiency: Millions spent per Goal scored
df_scout_finance['Cost_Per_Goal'] = df_scout_finance['Value_M'] / df_scout_finance['Goals']

print("Financial Scouting Summary:")
display(df_scout_finance[['Player', 'Goals', 'Cost_Per_Goal', 'Weekly_Wage_K']])
```
